```python
import pandas

df = pandas.read_csv("sqlite/dataset/comp3208-train.csv",names=['user_id',"item_id","rating","timestamp"])
df
```

```python
df=df.iloc[:,0:3].copy()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(15,5))
plt.style.use('ggplot')
x,y = np.unique(df["item_id"],return_counts=True)
y_sorted = np.argsort(-y)
plt.ylabel("Number of ratings")
plt.xlabel("Item_id")
plt.plot(x,y)
```

```python
plt.figure(figsize=(15,5))
plt.style.use('ggplot')
x,y = np.unique(df["user_id"],return_counts=True)
plt.ylabel("Number of ratings")
plt.xlabel("User_id")
plt.plot(x,y)
```
